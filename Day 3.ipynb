# Line plot

```python
month = ['Jan','Feb','Mar','Apr','May']
growth = [3.5,5,5.5,5.4,6]
```

```python
plt.plot(month,growth)
```

```python
plt.plot(month,growth,'o-')
```

```python
plt.plot(month,growth,'p--')
```

```python
plt.plot(month,growth,'o-',color='purple')
```

```python
plt.plot(month,growth,'o-',color='blue')
plt.xlabel('Month')
plt.ylabel('Growth in percentage')
```

```python
plt.plot(month,growth,'o-',color='blue')
plt.xlabel('Month')
plt.ylabel('Growth in percentage')
plt.title('Growth by Month')
plt.show()
```

# Bar Plot

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.species.unique()
```

```python
s=df.species.value_counts()
```

```python
plt.bar(x=s.index, height=s.values,color = ['blue','red','pink'])
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='species',data=df)  #barplot of a column
```

# Pie Chart

```python
s = df.species.value_counts()
plt.pie(labels=s.index,x=s.values,autopct="%.2f%%",colors=['red','green','blue'])
plt.show()
```

# Histogram
- We plot histogram for continuous numerical data
- Example: Temperature of a day, Height of a person

```python
plt.hist(df['sepal_length'])
```

```python
plt.hist(df['sepal_length'],bins=5)
```

```python
sns.histplot(df['petal_length'],kde=True) # histogram of petal length
```

# Box Plot
- For continuous numerical data

```python
sns.boxplot(df['sepal_width']) # Boxplot of petal length
```

# Scatterplot

```python
sns.scatterplot(x=df['petal_length'],y = df['petal_width'],hue=df['species'])
plt.show()
```

```python
from scipy.stats import pearsonr
r = pearsonr(df['petal_length'],df['petal_width'])
r.statistic
```
